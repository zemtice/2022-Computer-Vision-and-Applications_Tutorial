```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt 
```

```python
F=np.array([[1.4040868472652453E-7, 8.179088772122414E-6, -0.0031172345310651853],
            [1.1046936023622926E-6, 4.723322231237937E-11, -0.032528022287100104],
            [9.093794712467127E-5, 0.029573289609450402, 1.0]], dtype='float')
```

```python
left = cv.imread('left.jpeg')
right = cv.imread('right.jpeg')

left = cv.cvtColor(left, cv.COLOR_BGR2RGB)
right = cv.cvtColor(right, cv.COLOR_BGR2RGB)
print(left.shape)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Left')
plt.axis('off')
plt.imshow(left)
plt.subplot(1,2,2)
plt.title('Right')
plt.axis('off')
plt.imshow(right)
```

```python
left_ = left.copy()
right_ = right.copy()

pos_x = 0.55
pos_y = 0.55

pos_left = np.array([left_.shape[1]*pos_x, left_.shape[0]*pos_y, 1], dtype='float')

cv.circle(left_, (int(pos_left[0]), int(pos_left[1])), 4, (255,0,0), thickness=1)
line = F@pos_left
print(line)

for i in range(right_.shape[1]):
    x=i
    y=(-line[2]-line[0]*x)/line[1]
    if y<0 or y>=right_.shape[0]:
        continue
    right_[int(y),x]=[255,0,0]
    
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Left')
plt.axis('off')
plt.imshow(left_)
plt.subplot(1,2,2)
plt.title('Right')
plt.axis('off')
plt.imshow(right_)
    
```
